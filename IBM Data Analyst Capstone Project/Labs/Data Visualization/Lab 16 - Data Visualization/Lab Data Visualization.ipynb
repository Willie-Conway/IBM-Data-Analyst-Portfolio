**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import pandas as pd
import sqlite3

# Step 1: Load your CSV data into a DataFrame
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(file_path)

# Optional: check the first few rows
print(df.head())

# Step 2: Create a SQLite connection and database
conn = sqlite3.connect('survey-data.sqlite')

# Step 3: Write the DataFrame to a table called 'main'
df.to_sql('main', conn, if_exists='replace', index=False)

# Step 4: Close the connection
conn.close()

print("Data has been successfully written to survey-data.sqlite")
```

### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt  # <-- import matplotlib

# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run the query to get data for plotting
QUERY = "SELECT CompTotal FROM main"
df_comp_total = pd.read_sql_query(QUERY, conn)

# Plot a histogram of CompTotal
plt.hist(df_comp_total['CompTotal'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Total Compensation')
plt.xlabel('Total Compensation')
plt.ylabel('Frequency')
plt.show()

# Close the connection
conn.close()

```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here

# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run the query to get the 'Age' column
QUERY = "SELECT Age FROM main"
df_age = pd.read_sql_query(QUERY, conn)

# Define a mapping for age groups to midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': None  # Handle this case separately
}

# Map the age categories to their midpoints
df_age['Age_numeric'] = df_age['Age'].map(age_mapping)

# Check the mapping results
print(df_age[['Age', 'Age_numeric']].head())

# Drop rows where the 'Age_numeric' column is NaN
df_age_clean = df_age['Age_numeric'].dropna()

# Check if there are any valid values to plot
if not df_age_clean.empty:
    # Plot the box plot
    plt.boxplot(df_age_clean)
    plt.title('Box Plot of Age (Numeric Midpoints)')
    plt.ylabel('Age (Midpoint)')
    plt.show()
else:
    print("No valid numeric Age data to plot.")

# Close the connection
conn.close()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here

# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run the query to get data for plotting
QUERY = "SELECT Age, WorkExp FROM main"
df_scatter = pd.read_sql_query(QUERY, conn)

# Define a mapping for age groups to midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': None  # Handle this case separately
}

# Map the age categories to their midpoints
df_scatter['Age_numeric'] = df_scatter['Age'].map(age_mapping)

# Remove rows where 'Age_numeric' is NaN (e.g., 'Prefer not to say')
df_scatter_clean = df_scatter.dropna(subset=['Age_numeric'])

# Plot a scatter plot of Age vs WorkExp (using the numeric values of Age)
plt.scatter(df_scatter_clean['Age_numeric'], df_scatter_clean['WorkExp'], alpha=0.5, color='green')
plt.title('Scatter Plot of Age vs Work Experience')
plt.xlabel('Age (Numeric Midpoints)')
plt.ylabel('Work Experience')
plt.show()

# Close the connection
conn.close()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Fetch relevant columns
df = pd.read_sql_query("SELECT Age, TimeSearching, Frustration FROM main", conn)
conn.close()

# Map Age categories to numeric values for bubble size
age_map = {
    'Under 18 years old': 10,
    '18-24 years old': 20,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Map TimeSearching categories to approximate minutes
time_map = {
    'Less than 15 minutes a day': 7.5,
    '15-30 minutes a day': 22.5,
    '30-60 minutes a day': 45,
    '60-120 minutes a day': 90,
    '120+ minutes a day': 150
}
df['TimeSearchingNumeric'] = df['TimeSearching'].map(time_map)

# Drop rows with missing values
df_clean = df.dropna(subset=['TimeSearchingNumeric', 'Frustration', 'AgeNumeric'])

# Plot bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(
    df_clean['TimeSearchingNumeric'],
    df_clean['Frustration'],
    s=df_clean['AgeNumeric']*5,  # Bubble size scaled for visibility
    alpha=0.5,
    color='purple'
)
plt.title('Bubble Plot: Time Searching vs Frustration')
plt.xlabel('Time Searching (minutes)')
plt.ylabel('Frustration Level')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here

# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run the query to get the top 5 databases respondents want to learn
QUERY = """
SELECT DatabaseWantToWorkWith, COUNT(*) as count
FROM main
GROUP BY DatabaseWantToWorkWith
ORDER BY count DESC
LIMIT 5
"""
df_pie = pd.read_sql_query(QUERY, conn)

# Plot a pie chart
plt.pie(df_pie['count'], labels=df_pie['DatabaseWantToWorkWith'], autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Want to Learn')
plt.show()

# Close the connection
conn.close()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to database
conn = sqlite3.connect('survey-data.sqlite')

# Pull required columns
QUERY = """
SELECT Age, TimeSearching, TimeAnswering
FROM main
WHERE Age IN ('25-34 years old', '35-44 years old')
"""
df = pd.read_sql_query(QUERY, conn)
conn.close()

# Map Age to numeric
age_map = {
    '25-34 years old': 30,
    '35-44 years old': 35
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Map time ranges to numeric minutes
time_map = {
    'Less than 15 minutes a day': 7.5,
    '15-30 minutes a day': 22.5,
    '30-60 minutes a day': 45,
    '60-120 minutes a day': 90,
    'More than 120 minutes a day': 150
}

df['TimeSearching'] = df['TimeSearching'].map(time_map)
df['TimeAnswering'] = df['TimeAnswering'].map(time_map)

# Drop missing values
df = df.dropna(subset=['AgeNumeric', 'TimeSearching', 'TimeAnswering'])

# Calculate medians
df_median = df.groupby('AgeNumeric')[['TimeSearching', 'TimeAnswering']].median()

# Plot stacked bar chart
df_median.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Stacked Bar Chart: Median Time Searching vs Time Answering (Ages 30–35)')
plt.xlabel('Age')
plt.ylabel('Median Time (minutes)')
plt.legend(title='Activity')
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here

# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run the query to get median compensation for ages 45 to 60
QUERY = """
SELECT Age, AVG(CompTotal) as MedianCompTotal
FROM main
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
"""
df_line = pd.read_sql_query(QUERY, conn)

# Plot a line chart
plt.plot(df_line['Age'], df_line['MedianCompTotal'], marker='o', color='red')
plt.title('Line Chart: Median Total Compensation for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.show()

# Close the connection
conn.close()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here

# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run the query to get data for plotting
QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM main
GROUP BY MainBranch
ORDER BY count DESC
"""
df_bar = pd.read_sql_query(QUERY, conn)

# Plot a horizontal bar chart
df_bar.plot(kind='barh', x='MainBranch', y='count', color='skyblue', legend=False)
plt.title('Horizontal Bar Chart: Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()

# Close the connection
conn.close()
```

```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Fetch 'Age' and 'Employment' data
QUERY = "SELECT Age, Employment FROM main"
df_plot = pd.read_sql_query(QUERY, conn)

# Display the first 10 rows of the data to check for validity
print(df_plot.head(10))

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Handle missing data by replacing NaN values with 'Unknown'
df_plot['Age'] = df_plot['Age'].fillna('Unknown')
df_plot['Employment'] = df_plot['Employment'].fillna('Unknown')

# Split the 'Employment' column into multiple rows for each employment type
df_plot_expanded = df_plot.assign(Employment=df_plot['Employment'].str.split(';')).explode('Employment').reset_index(drop=True)

# Strip any leading or trailing whitespace from employment categories
df_plot_expanded['Employment'] = df_plot_expanded['Employment'].str.strip()

# Plotting a count plot with the expanded 'Employment' column
plt.figure(figsize=(10, 6))
sns.countplot(data=df_plot_expanded, x='Age', hue='Employment')
plt.title('Count Plot of Age vs Employment')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Close the connection
conn.close()
```
