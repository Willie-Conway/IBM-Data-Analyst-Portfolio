### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "survey-data.csv"
df = pd.read_csv(file_path)

# Clean the data: Convert 'ConvertedCompYearly' to numeric and drop rows with missing values in 'Age' or 'ConvertedCompYearly'
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df = df.dropna(subset=['ConvertedCompYearly', 'Age'])

# Group by 'Age' and calculate the median of 'ConvertedCompYearly' for each age group
age_group_medians = df.groupby('Age')['ConvertedCompYearly'].median()

# Plot a line chart of median compensation by age group
plt.figure(figsize=(10, 6))
plt.plot(age_group_medians.index, age_group_medians.values, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Add labels and title
plt.title('Median Yearly Compensation by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Yearly Compensation ($)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the chart
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "survey-data.csv"
df = pd.read_csv(file_path)

# Clean the data: Ensure 'ConvertedCompYearly' is numeric and handle missing values
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df = df.dropna(subset=['ConvertedCompYearly', 'Age'])

# Filter the dataset for ages between 25 and 45
age_filtered = df[df['Age'].isin(['25-34 years old', '35-44 years old'])]

# Group by 'Age' and calculate the median of 'ConvertedCompYearly'
age_group_medians = age_filtered.groupby('Age')['ConvertedCompYearly'].median()

# Plot a line chart for the median yearly compensation between ages 25-45
plt.figure(figsize=(10, 6))
plt.plot(age_group_medians.index, age_group_medians.values, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Add titles and labels
plt.title('Median Yearly Compensation for Ages 25 to 45', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Yearly Compensation ($)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the chart
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "survey-data.csv"
df = pd.read_csv(file_path)

# Clean the data: Ensure 'JobSatPoints_6' is numeric and handle missing values
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')
df = df.dropna(subset=['JobSatPoints_6', 'EdLevel'])

# Group by 'EdLevel' (education level) and calculate the average of 'JobSatPoints_6' (Job Satisfaction)
experience_group_job_sat = df.groupby('EdLevel')['JobSatPoints_6'].mean()

# Sort the values by Education Level for better plotting
experience_group_job_sat = experience_group_job_sat.sort_values()

# Plot a line chart for job satisfaction across different education levels
plt.figure(figsize=(12, 10))
plt.plot(experience_group_job_sat.index, experience_group_job_sat.values, marker='o', color='g', linestyle='-', linewidth=2, markersize=6)

# Add titles and labels
plt.title('Job Satisfaction by Experience Level (Education)', fontsize=16)
plt.xlabel('Experience Level (Education)', fontsize=12)
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "survey-data.csv"
df = pd.read_csv(file_path)

# Clean the data: Ensure 'ConvertedCompYearly' is numeric and handle missing values
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Drop rows with missing values in 'ConvertedCompYearly' and 'WorkExp' (or 'EdLevel' if 'WorkExp' is missing)
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'WorkExp'])

# Group by 'WorkExp' (or 'EdLevel' if 'WorkExp' is not available) and calculate the median compensation
experience_group_comp = df_cleaned.groupby('WorkExp')['ConvertedCompYearly'].median()

# Sort the values for better visualization
experience_group_comp = experience_group_comp.sort_values()

# Plot a line chart for median compensation across different experience levels
plt.figure(figsize=(12, 6))
plt.plot(experience_group_comp.index, experience_group_comp.values, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Add titles and labels
plt.title('Median Compensation by Experience Level', fontsize=16)
plt.xlabel('Experience Level (Years)', fontsize=12)
plt.ylabel('Median Compensation (ConvertedCompYearly)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "survey-data.csv"
df = pd.read_csv(file_path)

# Clean the data: Ensure 'JobSatPoints_6' is numeric and handle missing values
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows with missing values in 'JobSatPoints_6' and 'WorkExp' (or 'EdLevel' if 'WorkExp' is missing)
df_cleaned = df.dropna(subset=['JobSatPoints_6', 'WorkExp'])

# Group by 'WorkExp' (or 'EdLevel' if 'WorkExp' is not available) and calculate the mean job satisfaction
experience_group_sat = df_cleaned.groupby('WorkExp')['JobSatPoints_6'].mean()

# Sort the values for better visualization
experience_group_sat = experience_group_sat.sort_values()

# Plot a line chart for mean job satisfaction across different experience levels
plt.figure(figsize=(12, 6))
plt.plot(experience_group_sat.index, experience_group_sat.values, marker='o', color='g', linestyle='-', linewidth=2, markersize=6)

# Add titles and labels
plt.title('Job Satisfaction (JobSatPoints_6) Across Experience Levels', fontsize=16)
plt.xlabel('Experience Level (Years)', fontsize=12)
plt.ylabel('Mean Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.tight_layout()
plt.show()
```
