### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv("survey-data.csv")

# Step 3: Check the first few rows to understand the structure of the data
df.head()

# Step 4: Clean and preprocess the 'DatabaseWantToWorkWith' column
# Split the values in the 'DatabaseWantToWorkWith' column and count the frequency of each database
databases = df['DatabaseWantToWorkWith'].dropna().str.split(';', expand=True).stack()

# Count the frequency of each database
database_counts = databases.value_counts()

# Step 5: Select the top 5 most desired databases
top_5_databases = database_counts.head(5)

# Step 6: Create a pie chart for the top 5 databases
plt.figure(figsize=(8, 8))
top_5_databases.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#ff6666','#c2c2f0'])
plt.title('Top 5 Databases Respondents Want to Work With', fontsize=14)
plt.ylabel('')  # To remove the 'y' label on the left side
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv("survey-data.csv")

# Step 3: Clean and preprocess the 'LanguageHaveWorkedWith' column
# Since 'LanguageHaveWorkedWith' is a semi-colon separated string, we need to split and count the occurrences of each language.
language_series = df['LanguageHaveWorkedWith'].dropna().str.split(';', expand=True).stack()

# Step 4: Count the occurrences of each language
language_counts = language_series.value_counts()

# Step 5: Select the top 5 most common programming languages
top_5_languages = language_counts.head(5)

# Step 6: Create a pie chart for the top 5 programming languages respondents have worked with
plt.figure(figsize=(8, 8))
top_5_languages.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#ff6666','#c2c2f0'])
plt.title('Top 5 Programming Languages Respondents Have Worked With', fontsize=14)
plt.ylabel('')  # Remove the y-axis label
plt.show()
```
