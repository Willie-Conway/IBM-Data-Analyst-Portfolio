#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Plot the scatter plot for Age vs Job Satisfaction
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['JobSatPoints_6'], alpha=0.5, color='blue')

# Adding labels and title
plt.title('Scatter Plot: Age vs Job Satisfaction', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Improve readability of the x-axis
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.tight_layout()  # Adjust layout to avoid overlap of labels

# Show the plot with grid
plt.grid(True)
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
!pip install seaborn
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # For regression line

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Map categorical 'Age' to midpoints as we did before
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 65,
    'Prefer not to say': None  # Treat this as missing data
}

# Replace age categories with midpoints
df['Age'] = df['Age'].map(age_mapping)

# Convert 'JobSatPoints_6' to numeric and handle errors by coercing invalid values to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Now drop rows where either 'Age' or 'JobSatPoints_6' is NaN (missing)
df_cleaned = df.dropna(subset=['Age', 'JobSatPoints_6'])

# Check how many missing values there are before and after cleaning
print(f"Missing values in 'Age' before cleaning: {df['Age'].isna().sum()}")
print(f"Missing values in 'JobSatPoints_6' before cleaning: {df['JobSatPoints_6'].isna().sum()}")
print(f"Rows remaining after cleaning: {len(df_cleaned)}")

# Create the scatter plot with regression line
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['Age'], df_cleaned['JobSatPoints_6'], alpha=0.5, color='blue')

# Add regression line using seaborn (without scatter, as it's already plotted)
sns.regplot(x='Age', y='JobSatPoints_6', data=df_cleaned, scatter=False, line_kws={'color': 'red', 'lw': 2})

# Adding labels and title
plt.title('Scatter Plot with Trend Line: Age vs Job Satisfaction', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Improve readability of the x-axis
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.tight_layout()  # Adjust layout to avoid overlap of labels

# Show the plot with grid
plt.grid(True)
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Map categorical age ranges to midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 65,
    'Prefer not to say': None  # We can treat this as missing data
}

# Replace age categories with midpoints (or NaN for 'Prefer not to say')
df['Age'] = df['Age'].map(age_mapping)

# Check for missing values in 'Age' and 'YearsCodePro' after mapping
print(f"Missing values in 'Age' after mapping: {df['Age'].isna().sum()}")
print(f"Missing values in 'YearsCodePro' after mapping: {df['YearsCodePro'].isna().sum()}")

# Clean 'YearsCodePro' column, making sure its numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Impute missing 'Age' and 'YearsCodePro' with their respective median values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['YearsCodePro'] = df['YearsCodePro'].fillna(df['YearsCodePro'].median())

# Check for missing values after imputation
print(f"Missing values in 'Age' after filling: {df['Age'].isna().sum()}")
print(f"Missing values in 'YearsCodePro' after filling: {df['YearsCodePro'].isna().sum()}")

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['YearsCodePro'], alpha=0.5, color='blue')

# Adding labels and title
plt.title('Scatter Plot: Age vs Work Experience', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Work Experience (YearsCodePro)', fontsize=12)

# Improve readability of the x-axis
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.tight_layout()  # Adjust layout to avoid overlap of labels

# Show the plot with grid
plt.grid(True)
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Normalize Age to make the bubble size more visually appealing
# We scale the Age by a factor of 100 to make it more noticeable
df_cleaned['Age_scaled'] = df_cleaned['Age'] * 100

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(
    x=df_cleaned['ConvertedCompYearly'], 
    y=df_cleaned['JobSatPoints_6'], 
    s=df_cleaned['Age_scaled'],  # Bubble size based on Age
    alpha=0.5,                  # Transparency of the bubbles
    c=df_cleaned['Age'],        # Color of bubbles based on Age
    cmap='viridis',             # Colormap for the bubble color
    edgecolors="w",             # White border around bubbles
    linewidth=0.5               # Line width for the border
)

# Add labels and title
plt.title('Bubble Plot: Compensation vs Job Satisfaction with Age as Bubble Size', fontsize=14)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Add a color bar to represent the age
plt.colorbar(label='Age')

# Show the plot with grid
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Ensure JobSatPoints_6 is numeric
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows with missing JobSatPoints_6 or LanguageHaveWorkedWith
df_cleaned = df.dropna(subset=['JobSatPoints_6', 'LanguageHaveWorkedWith'])

# Split the languages into multiple rows
df_languages = df_cleaned.assign(Language=df_cleaned['LanguageHaveWorkedWith'].str.split(';')).explode('Language')

# Strip whitespace from language names
df_languages['Language'] = df_languages['Language'].str.strip()

# Calculate average job satisfaction per language
language_satisfaction = df_languages.groupby('Language')['JobSatPoints_6'].mean().sort_values(ascending=False)

# Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(
    x=language_satisfaction.index,
    y=language_satisfaction.values,
    s=100,          # Marker size
    color='skyblue',
    edgecolor='black'
)

# Labels and title
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.xlabel('Programming Language', fontsize=12)
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Programming Languages vs Job Satisfaction', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Convert numeric columns
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows with missing values in key columns
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Employment'])

# Keep only the top 5 employment types (most common) for clarity
top_employment_types = df_cleaned['Employment'].value_counts().nlargest(5).index
df_cleaned = df_cleaned[df_cleaned['Employment'].isin(top_employment_types)]

# Create the scatter plot
plt.figure(figsize=(12, 6))

# Plot each employment type with a different color
for employment_type in top_employment_types:
    subset = df_cleaned[df_cleaned['Employment'] == employment_type]
    plt.scatter(
        subset['ConvertedCompYearly'],
        subset['JobSatPoints_6'],
        alpha=0.5,
        label=employment_type
    )

# Labels and title
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.title('Compensation vs Job Satisfaction by Employment Type', fontsize=14)
plt.legend(title='Employment Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Convert YearsCodePro to numeric (non-numeric values become NaN)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop rows with missing YearsCodePro, Age, or Country
df_cleaned = df.dropna(subset=['YearsCodePro', 'Age', 'Country'])

# Optionally, keep only top 5 countries for clarity
top_countries = df_cleaned['Country'].value_counts().nlargest(5).index
df_cleaned = df_cleaned[df_cleaned['Country'].isin(top_countries)]

# Define marker shapes for age groups
age_groups = df_cleaned['Age'].unique()
markers = ['o', 's', '^', 'D', 'v', 'P', '*', 'X']  # enough marker types

marker_map = {age: markers[i % len(markers)] for i, age in enumerate(age_groups)}

# Plot
plt.figure(figsize=(12, 6))
for age in age_groups:
    subset = df_cleaned[df_cleaned['Age'] == age]
    plt.scatter(
        subset['Country'],
        subset['YearsCodePro'],
        marker=marker_map[age],
        alpha=0.6,
        label=age
    )

# Labels and title
plt.xlabel('Country', fontsize=12)
plt.ylabel('Years of Professional Coding (YearsCodePro)', fontsize=12)
plt.title('Work Experience vs Age Group by Country', fontsize=14)

# Show legend for age groups
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
