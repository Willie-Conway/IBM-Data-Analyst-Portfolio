### Step 2: Load the Dataset into a DataFrame



load the dataset using pd.read_csv()


```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3: Identifying Duplicate Rows


**Task 1: Identify Duplicate Rows**
  1. Count the number of duplicate rows in the dataset.
  2. Display the first few duplicate rows to understand their structure.


```python
## Write your code here

# Count the number of duplicate rows
duplicate_count = df.duplicated().sum()

# Display the first few duplicate rows
duplicates = df[df.duplicated()]

print(f"Number of duplicate rows: {duplicate_count}")
print("\nFirst few duplicate rows:")
print(duplicates.head())
```

### Step 4: Removing Duplicate Rows


**Task 2: Remove Duplicates**
   1. Remove duplicate rows from the dataset using the drop_duplicates() function.
2. Verify the removal by counting the number of duplicate rows after removal .


```python
## Write your code here

# Remove duplicate rows
df_cleaned = df.drop_duplicates()

# Verify removal by counting the number of duplicate rows after removal
duplicate_count_after_removal = df_cleaned.duplicated().sum()

print(f"Number of duplicate rows after removal: {duplicate_count_after_removal}")
```

### Step 5: Handling Missing Values


**Task 3: Identify and Handle Missing Values**
   1. Identify missing values for all columns in the dataset.
   2. Choose a column with significant missing values (e.g., EdLevel) and impute with the most frequent value.


```python
## Write your code here

# Check for missing values in all columns
missing_values = df_cleaned.isnull().sum()

print("Missing values per column:")
print(missing_values)

# Handle missing values in 'EdLevel' by imputing with the most frequent value
most_frequent_value = df_cleaned['EdLevel'].mode()[0]  # Get the most frequent value
df_cleaned['EdLevel'] = df_cleaned['EdLevel'].fillna(most_frequent_value)

# Verify that there are no missing values in 'EdLevel' anymore
missing_values_after_imputation = df_cleaned['EdLevel'].isnull().sum()

print(f"Missing values in 'EdLevel' after imputation: {missing_values_after_imputation}")
```

### Step 6: Normalizing Compensation Data


**Task 4: Normalize Compensation Data Using ConvertedCompYearly**
   1. Use the ConvertedCompYearly column for compensation analysis as the normalized annual compensation is already provided.
   2. Check for missing values in ConvertedCompYearly and handle them if necessary.


```python
## Write your code here

# Check for missing values in the 'ConvertedCompYearly' column
missing_compensation = df_cleaned['ConvertedCompYearly'].isnull().sum()

print(f"Missing values in 'ConvertedCompYearly': {missing_compensation}")

# If there are missing values, impute them with the mean value (or use another strategy)
if missing_compensation > 0:
    mean_compensation = df_cleaned['ConvertedCompYearly'].mean()
    df_cleaned['ConvertedCompYearly'] = df_cleaned['ConvertedCompYearly'].fillna(mean_compensation)

# Verify that there are no missing values in 'ConvertedCompYearly' anymore
missing_compensation_after_imputation = df_cleaned['ConvertedCompYearly'].isnull().sum()

print(f"Missing values in 'ConvertedCompYearly' after imputation: {missing_compensation_after_imputation}")
```
