### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here

# Display column names, data types, and summary information
df.info()

# Summary statistics
df.describe(include='all')
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here

# Analyze the distribution of values in key columns
df['Employment'].value_counts(), df['JobSat'].value_counts(), df['YearsCodePro'].value_counts()
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


### Pie Chart

```python
## Write your code here

# Display the first 8 most common job satisfaction values
job_sat_counts = df['JobSat'].value_counts().head(8)

# Pie chart for the first 8 job satisfaction distributions
plt.figure(figsize=(8, 8))
job_sat_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title('Top 8 Job Satisfaction Distribution')
plt.ylabel('')
plt.show()
```

### KDE Plot

```python
# KDE plot for Job Satisfaction with the 'fill' parameter
sns.kdeplot(df['JobSat'].dropna(), fill=True)
plt.title('Job Satisfaction Distribution (KDE)')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here

# Analyzing the programming languages data (Top 10 languages)
languages_worked_with = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode().value_counts().head(10)
languages_want_to_work_with = df['LanguageWantToWorkWith'].dropna().str.split(';').explode().value_counts().head(10)

# Visualizing the overlap with a grouped bar chart
languages_combined = pd.DataFrame({
    'Worked With': languages_worked_with,
    'Want to Work With': languages_want_to_work_with
}).fillna(0)

# Plotting the bar chart for the top 10 languages
languages_combined.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Programming Languages: Worked With vs Want to Work With')
plt.ylabel('Count')
plt.xlabel('Programming Language')
plt.xticks(rotation=45)
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

# Assuming 'RemoteWork' column is present and contains information about remote work preferences
region_remote_work = pd.crosstab(df['RemoteWork'], df['Country'], normalize='index')

# Visualize the distribution using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(region_remote_work, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Remote Work Distribution by Region')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsCodePro', y='JobSat', data=df)
plt.title('Job Satisfaction vs. Years of Professional Coding Experience')
plt.show()
```

###  Check for Missing Values in the Relevant Columns

```python
# Check for missing values in the relevant columns
print(df[['YearsCodePro', 'JobSat']].isnull().sum())
```

### Clean the Data (Remove Rows with Missing or Invalid Values)

```python
# Drop rows with missing values in the columns 'YearsCodePro' or 'JobSat'
df_cleaned = df.dropna(subset=['YearsCodePro', 'JobSat'])

# Ensure that 'YearsCodePro' is numeric and handle any other potential issues
df_cleaned['YearsCodePro'] = pd.to_numeric(df_cleaned['YearsCodePro'], errors='coerce')

# Drop any rows with 'NaN' values after converting 'YearsCodePro' to numeric
df_cleaned = df_cleaned.dropna(subset=['YearsCodePro', 'JobSat'])

# Now plot the scatter plot again for the cleaned data
plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsCodePro', y='JobSat', data=df_cleaned)
plt.title('Job Satisfaction vs. Years of Professional Coding Experience (Cleaned Data)')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction')
plt.show()
```

### If Your Data Contains Zero Variance (Constant Columns)

```python
# Check the variance of 'YearsCodePro' to ensure there's variation
print(df_cleaned['YearsCodePro'].var())
```

```python
# Map YearsCodePro to numeric values
years_mapping = {
    'Less than 1 year': 0,  # Assuming 'Less than 1 year' means 0 years of experience
    '1-2 years': 1.5,       # Mid-point of the range 1-2 years
    '3-5 years': 4,         # Mid-point of the range 3-5 years
    '6-10 years': 8,        # Mid-point of the range 6-10 years
    '11-15 years': 13,      # Mid-point of the range 11-15 years
    '16-20 years': 18,      # Mid-point of the range 16-20 years
    '21-25 years': 23,      # Mid-point of the range 21-25 years
    '26-30 years': 28,      # Mid-point of the range 26-30 years
    '30+ years': 30         # Assuming '30+ years' means 30 years of experience
}

# Convert YearsCodePro to numeric values
df['YearsCodePro'] = df['YearsCodePro'].map(years_mapping)

# Drop rows with missing values in 'YearsCodePro' or 'JobSat'
df_cleaned = df.dropna(subset=['YearsCodePro', 'JobSat'])

# Check the variance of the 'YearsCodePro' column
variance_years_code_pro = df_cleaned['YearsCodePro'].var()
print(f"Variance of 'YearsCodePro': {variance_years_code_pro}")

# If variance is zero, correlation cannot be calculated
if variance_years_code_pro != 0:
    # Calculate Pearson correlation coefficient
    correlation = df_cleaned['JobSat'].corr(df_cleaned['YearsCodePro'])
    print(f'Pearson Correlation between Job Satisfaction and Years of Professional Coding Experience: {correlation}')
else:
    print("Cannot calculate correlation because 'YearsCodePro' has zero variance.")
```

```python
# Drop rows with missing values in either 'JobSat' or 'YearsCodePro'
df_cleaned = df.dropna(subset=['JobSat', 'YearsCodePro'])

# Recalculate the correlation on the cleaned data
correlation = df_cleaned['JobSat'].corr(df_cleaned['YearsCodePro'])
print(f'Pearson Correlation between Job Satisfaction and Years of Professional Coding Experience: {correlation}')
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here

# Cross-tabulation of Employment and Education Level
employment_vs_edlevel = pd.crosstab(df['Employment'], df['EdLevel'])

# Show the first 10 rows of the cross-tabulation
print(employment_vs_edlevel.head(10))

# Visualize with a stacked bar plot
employment_vs_edlevel.head(10).plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Employment Status vs. Education Level (First 10 Rows)')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.show()
```
