## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
# Step 1: Import the required libraries
import requests
from bs4 import BeautifulSoup
import csv
```

Download the webpage at the url


```python
#your code goes here
# Step 2: Download the webpage at the URL
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
response = requests.get(url)
```

Create a soup object


```python
#your code goes here
# Step 3: Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
# Step 4: Scrape the language name and annual average salary
# Assuming the structure of the table is consistent and the language and salary are in the same row
languages_data = []
```

Find the table containing the data

```python
# Find the table containing the data
table = soup.find('table', {'class': 'table'})
```

Find all the rows in the table (skip the header row)

```python
# Print the first 1000 characters of the soup to see the HTML structure
print(soup.prettify()[:1000])
```

```python
# Find the table containing the data (assuming it's the first table in the document)
table = soup.find('table')  # Find the first <table> tag
rows = table.find_all('tr')[1:]  # Skip the first row (header)
```

Loop through each row to extract the data

```python
# Loop through each row to extract the data
for row in rows:
    cols = row.find_all('td')  # Find all columns in the row
    if len(cols) > 1:  # Ensure the row has enough columns
        language = cols[1].text.strip()  # The language is in the second column (index 1)
        salary = cols[3].text.strip()  # The salary is in the fourth column (index 3)
        languages_data.append([language, salary])
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
# Step 5: Save the scraped data into a CSV file
with open('popular-languages.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Programming Language', 'Average Annual Salary'])  # Write header
    writer.writerows(languages_data)  # Write the data

print("Data has been successfully scraped and saved to 'popular-languages.csv'")
```
