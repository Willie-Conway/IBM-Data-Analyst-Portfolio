Print the mean age of the survey participants.


```python
## Write your code here
# Write your code here
# Write your code here
import pandas as pd
import re
import numpy as np

# Function to extract numeric age or midpoint from age ranges
def extract_age(age_str):
    if pd.isna(age_str):
        return np.nan
    age_str = str(age_str).strip()
    
    # If it's a range like '25-34'
    if '-' in age_str:
        parts = re.findall(r'\d+', age_str)
        if len(parts) == 2:
            return (int(parts[0]) + int(parts[1])) / 2
    # If it's a single number like '44 years old'
    match = re.search(r'\d+', age_str)
    if match:
        return int(match.group())
    
    return np.nan

# Apply conversion
df['Age_numeric'] = df['Age'].apply(extract_age)

# Calculate and print mean age
mean_age = df['Age_numeric'].mean()
print("Mean age of participants:", mean_age)
```
